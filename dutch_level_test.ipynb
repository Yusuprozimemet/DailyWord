```python
import random

class DutchLanguageTest:
    def __init__(self):
        self.questions = {
            'A1': [
                {'question': 'Wat is "hello" in het Nederlands?', 'options': ['Hallo', 'Bonjour', 'Ciao', 'Hi'], 'correct': 'Hallo'},
                {'question': 'Hoe zeg je "thank you"?', 'options': ['Alsjeblieft', 'Dankjewel', 'Graag', 'Dag'], 'correct': 'Dankjewel'},
                {'question': 'Wat betekent "tot ziens"?', 'options': ['See you later', 'Goodbye', 'Thank you', 'Hello'], 'correct': 'Goodbye'},
                {'question': 'Wat is de kleur van de lucht op een heldere dag?', 'options': ['Blauw', 'Groen', 'Geel', 'Rood'], 'correct': 'Blauw'},
                {'question': 'Hoe zeg je "good morning"?', 'options': ['Goedenacht', 'Goedemiddag', 'Goedemorgen', 'Hallo'], 'correct': 'Goedemorgen'},
                {'question': 'Wat betekent "ik ben blij"?', 'options': ['I am sad', 'I am happy', 'I am angry', 'I am tired'], 'correct': 'I am happy'},
                {'question': 'Wat is de Nederlandse vertaling van "apple"?', 'options': ['Peer', 'Appel', 'Banaan', 'Sinaasappel'], 'correct': 'Appel'},
                {'question': 'Wat zeg je als je iemand groet in de avond?', 'options': ['Goedemorgen', 'Goedenavond', 'Dag', 'Doeg'], 'correct': 'Goedenavond'},
                {'question': 'Welke dag is het na zondag?', 'options': ['Vrijdag', 'Dinsdag', 'Maandag', 'Zaterdag'], 'correct': 'Maandag'},
                {'question': 'Wat betekent "fiets"?', 'options': ['Bike', 'Car', 'Train', 'Plane'], 'correct': 'Bike'}
            ],
            'A2': [
                {'question': 'Welk woord past in de zin: "Ik ___ naar school"?', 'options': ['ga', 'gaat', 'gaan', 'ging'], 'correct': 'ga'},
                {'question': 'Wat is het meervoud van "huis"?', 'options': ['huisen', 'huizen', 'huisse', 'huis'], 'correct': 'huizen'},
                {'question': 'Hoe zeg je "I am eating"?', 'options': ['Ik eet', 'Ik eette', 'Ik eten', 'Ik gegeten'], 'correct': 'Ik eet'},
                {'question': 'Welke dag komt na maandag?', 'options': ['Dinsdag', 'Woensdag', 'Donderdag', 'Vrijdag'], 'correct': 'Dinsdag'},
                {'question': 'Wat is het tegenovergestelde van "groot"?', 'options': ['Klein', 'Lang', 'Dik', 'Kort'], 'correct': 'Klein'},
                {'question': 'Wat betekent "ik wil graag een kop koffie"?', 'options': ['I want a cup of coffee', 'I need some milk', 'I am tired', 'I am happy'], 'correct': 'I want a cup of coffee'},
                {'question': 'Wat is het verleden tijd van "lopen"?', 'options': ['Liepen', 'Liep', 'Loopte', 'Lopen'], 'correct': 'Liep'},
                {'question': 'Wat betekent "Ik houd van boeken"?', 'options': ['I like books', 'I read books', 'I write books', 'I buy books'], 'correct': 'I like books'},
                {'question': 'Wat zeg je als je iemand uitnodigt?', 'options': ['Wil je komen?', 'Tot ziens!', 'Bedankt', 'Geen probleem'], 'correct': 'Wil je komen?'},
                {'question': 'Wat is een synoniem van "vriend"?', 'options': ['Maat', 'Vijand', 'Meisje', 'Familie'], 'correct': 'Maat'}
            ],
            'B1': [
                {'question': 'Welke zin is grammaticaal correct?', 'options': [
                    'Ik heb gisteren naar de film gekeken',
                    'Ik heb gisteren naar de film gekijkt',
                    'Ik ben gisteren naar de film gekeken',
                    'Ik ben gisteren de film gekeken'
                ], 'correct': 'Ik heb gisteren naar de film gekeken'},
                {'question': 'Wat betekent "Als ik tijd had, zou ik komen"?', 'options': [
                    'I will come if I have time',
                    'I would come if I had time',
                    'I have time to come',
                    'I came because I had time'
                ], 'correct': 'I would come if I had time'},
                {'question': 'Wat is het juiste voltooid deelwoord van "lopen"?', 'options': ['gelopen', 'loopte', 'gelopenen', 'gelopen'], 'correct': 'gelopen'},
                {'question': 'Wat is een synoniem van "moeilijk"?', 'options': ['Lastig', 'Makkelijk', 'Simpel', 'Eenvoudig'], 'correct': 'Lastig'},
                {'question': 'Wat is een correcte vertaling van "I have been waiting for an hour"?', 'options': [
                    'Ik wacht een uur',
                    'Ik ben al een uur aan het wachten',
                    'Ik heb gewacht voor een uur',
                    'Ik was aan het wachten voor een uur'
                ], 'correct': 'Ik ben al een uur aan het wachten'},
                {'question': 'Wat betekent "Ik ga naar de bakker om brood te kopen"?', 'options': [
                    'I go to the baker to buy bread',
                    'I buy bread for the baker',
                    'I bake bread for the baker',
                    'I wait at the bakery'
                ], 'correct': 'I go to the baker to buy bread'},
                {'question': 'Welke zin bevat een fout?', 'options': [
                    'Wij hebben gisteren een taart gebakken',
                    'Ik gaat naar de winkel',
                    'Hij leest een boek',
                    'Zij heeft haar werk afgemaakt'
                ], 'correct': 'Ik gaat naar de winkel'},
                {'question': 'Wat is een antoniem van "snel"?', 'options': ['Langzaam', 'Vlug', 'Behendig', 'Krachtig'], 'correct': 'Langzaam'},
                {'question': 'Hoe zeg je "She is speaking Dutch very well"?', 'options': [
                    'Zij spreekt heel goed Nederlands',
                    'Zij heeft goed Nederlands praten',
                    'Zij praat Nederlands heel goed',
                    'Zij zijn goed Nederlands aan het spreken'
                ], 'correct': 'Zij spreekt heel goed Nederlands'},
                {'question': 'Wat betekent "Ik begrijp het niet"?', 'options': ['I don’t understand it', 'I don’t need it', 'I am not buying it', 'I don’t want it'], 'correct': 'I don’t understand it'}
            ]
        }
        self.user_score = 0
        self.total_questions = 0


    def shuffle_questions(self):
        """Shuffle questions and options for each level."""
        for level, questions in self.questions.items():
            random.shuffle(questions)
            for question in questions:
                random.shuffle(question['options'])

    def ask_question(self, question):
        """Ask a single question and return whether the user answered correctly."""
        print(f"\n{question['question']}")
        for idx, option in enumerate(question['options'], 1):
            print(f"{idx}. {option}")

        while True:
            try:
                choice = int(input("\nKies het nummer van je antwoord (1-4): "))
                if 1 <= choice <= len(question['options']):
                    return question['options'][choice - 1] == question['correct']
                print("Ongeldige keuze. Kies een nummer tussen 1 en 4.")
            except ValueError:
                print("Voer een geldig nummer in.")

    def run_test(self):
        """Run the language test."""
        print("Welkom bij de Nederlandse Taaltest!")
        print("====================================")

        self.shuffle_questions()  # Shuffle questions before starting

        for level, questions in self.questions.items():
            print(f"\nLevel {level} vragen:")
            for question in questions:
                self.total_questions += 1
                if self.ask_question(question):
                    self.user_score += 1
                    print("Correct!")
                else:
                    print(f"Helaas, het juiste antwoord was: {question['correct']}")

    def calculate_level(self):
        """Calculate the user's estimated Dutch language level based on their score."""
        percentage = (self.user_score / self.total_questions) * 100
        if percentage >= 80:
            return "B1/B2"
        elif percentage >= 60:
            return "A2"
        return "A1"

    def show_results(self):
        """Display the test results."""
        print("\n====================================")
        print("Testresultaten:")
        print(f"Totaal correct: {self.user_score}/{self.total_questions}")
        print(f"Percentage: {(self.user_score / self.total_questions) * 100:.1f}%")
        print(f"Geschat niveau: {self.calculate_level()}")
        print("====================================")


```

```python
def main():
    test = DutchLanguageTest()
    test.run_test()
    test.show_results()

if __name__ == "__main__":
    main()
```
